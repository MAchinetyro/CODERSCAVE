```python
#Importing all the  Libraries
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading Dataset
dataset=pd.read_csv('globalterrorismdb_0718dist.csv',encoding="ISO-8859-1")
```

```python
#Cleaning Data
dataset=dataset[['iyear','imonth','iday','country_txt','region_txt','gname','provstate','city','latitude','longitude','location','summary',
      'attacktype1_txt','targtype1_txt','weaptype1_txt','nkill','nwound']]
```

```python
dataset.head()
```

```python
dataset.columns.nunique()
```

```python
missing_counts=dataset.isnull().sum()
print(missing_counts)
```

```python
attacks=dataset["iyear"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")
attacks.head()
```

```python
print("COUNTRY WITH THE MOST ATTACKS:",dataset['country_txt'].value_counts().idxmax())
print("YEAR WITH THE MOST ATTACKS:",dataset['iyear'].value_counts().idxmax())
print("REGION WITH THE MOST ATTACKS:",dataset['region_txt'].value_counts().idxmax())
print("GROUP WITH THE MOST ATTACKS:",dataset['gname'].value_counts().index[1])
print("STATE WITH THE MOST ATTACKS",dataset["provstate"].value_counts().idxmax())
print("MONTH WITH THE MOST ATTACKS:",dataset['imonth'].value_counts().idxmax())
```

#Timeline of attack over the years


```python
attacks.plot(kind="bar",color="Yellow",figsize=(15,6),fontsize=13)
plt.title("Timeline of Attacks",fontsize=15)
plt.xlabel("Years",fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.show()
```

```python
#Total killed in each Year
AnnualyKilled=dataset[['iyear','nkill']].groupby('iyear').sum()
AnnualyKilled.head()
```

```python
#Total wounded in each year
annualwounded=dataset[['iyear','nwound']].groupby('iyear').sum()
annualwounded.head()
```

```python
fig=plt.figure()
ax0=fig.add_subplot(2,1,1)
#Visualisation of People-Killed in each year
AnnualyKilled.plot(kind="bar",color="Orange",figsize=(15,14),ax=ax0)
ax0.set_title("People Killed in Each Year")
ax0.set_xlabel("YEAR")
ax0.set_ylabel("No Of People Killed")
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,1,2)
#Visualization of People Wounded in Each Year
annualwounded.plot(kind="bar",color="aquamarine",figsize=(15,14),ax=ax1)
ax1.set_title("People Wounded In Each Year")
ax1.set_xlabel("YEAR")
ax1.set_ylabel("Number Of People Wounded")
```

```python
reg=pd.crosstab(dataset.iyear,dataset.region_txt)
reg.head()
```

```python
#Region-wise attack visualisation
reg.plot(kind="area", stacked=False, alpha=0.5,figsize=(20,10))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

Correlation Analysis


```python
sns.heatmap(np.round(dataset.corr(),2),annot=True, cmap="BuPu")
```

```python

top_10_countries = dataset['country_txt'].value_counts().head(10)

plt.figure(figsize=(15, 10))
sns.barplot(x=top_10_countries.index, y=top_10_countries.values, palette='YlOrBr_r')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

```

```python

dataset['nwound'] = dataset['nwound'].fillna(0).astype(int)
dataset['nkill'] = dataset['nkill'].fillna(0).astype(int)
dataset['casualities'] = dataset['nkill'] + dataset['nwound']

terrorism = dataset.sort_values(by='casualities',ascending=False)[:40]
feat = terrorism.pivot_table(index='country_txt',columns='iyear',values='casualities')
feat.fillna(0,inplace=True)
```

```python
feat.head()
```

```python

plt.figure(figsize=(15, 6))
sns.barplot(x=dataset['country_txt'].value_counts()[:15].index, y=dataset['country_txt'].value_counts()[:15].values, palette='Blues_d')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
#Country-wise killed by each group
kcg=dataset[["gname","country_txt","nkill"]].groupby(["gname","country_txt"],axis=0).sum().sort_values(by="nkill",ascending=False).drop("Unknown").head(10)
kcg
```

```python
#Countries with most people killed by each group
kcg.plot(kind="bar",color="violet",figsize=(15,6))
plt.title("Countries with most people killed by a particular Group",fontsize=13)
plt.xlabel("(Terrorist Group,Country)",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of people Killed",fontsize=13)
plt.show()
```
